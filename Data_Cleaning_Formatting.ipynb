```python
import path
import glob
import os
from datetime import datetime
import pandas as pd
```

```python
#Loop thru all CSV files, and add to DataFrame

path = 'CityBikeData'

Files = glob.glob(path + "/*.csv")

df = pd.concat([pd.read_csv(f, index_col=None, header=0, parse_dates=['starttime'], 
    date_parser=lambda x: pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S')) for f in Files])

df

```
